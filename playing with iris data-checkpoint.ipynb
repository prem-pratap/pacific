```python
import pandas as pd
import seaborn as sb
from sklearn.datasets import load_iris 
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("C:/Users/DBMS/Desktop/datasets/iris.csv")
iris=load_iris()
```

```python
df.info()
print(df.columns)
print(df.shape)
#dir(iris)

print(iris.feature_names)
print(iris.data.shape)
```

```python
sb.distplot(a=df["sepal_length"])#kde=true & hist=true by default 
plt.show()
```

```python
#density of kernel is find by facetgrid
'''
a final seaborn plot useful  for looking at univariate relations is the kdeplot
which creates and visualizes a kernel density estimate of the underlying features
not univariate in complete sense.3 different plots for 3 species values
\ it is used for continuation
'''
sb.FacetGrid(df,hue="type",size=4) \
    .map(sb.kdeplot , "petal_length") \
    .add_legend()
plt.show()
```

```python
#
sb.factorplot(x="sepal_length",y="sepal_length",data=df,hue="type",size=5)
sb.factorplot(x="sepal_length",y="petal_length",data=df,hue="type",size=5)
sb.factorplot(x="petal_length",y="sepal_length",data=df,hue="type",size=5)
sb.factorplot(x="petal_length",y="petal_length",data=df,hue="type",size=5)
#a seaborn jointplot shows bivariate scatterplots and univariate histogram
sb.jointplot(x="sepal_length",y="sepal_width",data=df,size=5,kind="scatter")
plt.show()
```
